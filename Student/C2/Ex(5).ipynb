### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv). 

### Step 3. Assign it to a variable called euro12.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep="|")
```

### Step 4. Select only the Goal column.

```python
users.groupby("occupation")["age"].mean()
```

### Step 5. How many team participated in the Euro2012?

```python
users["is_male"] = users["gender"].apply(lambda x: 1 if x == "M" else 0)

users.groupby("occupation")["is_male"].mean().sort_values(ascending=False)
```

### Step 6. What is the number of columns in the dataset?

```python
users.groupby("occupation")["age"].agg(["min", "max"])
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
users.groupby(["occupation", "gender"])["age"].mean()
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
gender_count = users.groupby(["occupation", "gender"]).size()
total = users.groupby("occupation").size()

gender_percent = gender_count.div(total, level="occupation") * 100
gender_percent
```
