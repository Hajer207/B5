## Introduction

```python
import pandas as pd
```

```python
df = pd.read_csv("WA_Marketing-Campaign.csv")
df.head()
```

```python
df.groupby("Promotion")["SalesInThousands"].mean()
```

## Results

The average sales for each promotion are as follows:

- Promotion 1: 58.09
- Promotion 2: 47.33
- Promotion 3: 55.36

Promotion 1 achieved the highest average sales among all campaigns.

```python
group1 = df[df["Promotion"] == 1]["SalesInThousands"]
group2 = df[df["Promotion"] == 2]["SalesInThousands"]
group3 = df[df["Promotion"] == 3]["SalesInThousands"]

from scipy import stats
stats.f_oneway(group1, group2, group3)
```

## Statistical Test (ANOVA)

The ANOVA test results show a p-value of 6.76e-10, which is less than 0.05.

This indicates that there is a statistically significant difference in sales between the three promotions.

```python
stats.f_oneway(group1, group2, group3)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(
    endog=df["SalesInThousands"],
    groups=df["Promotion"],
    alpha=0.05
)

print(tukey)
```

### Post-hoc Test (Tukey HSD)

The Tukey HSD test results show the following:

- There is a significant difference between Promotion 1 and Promotion 2.
- There is NO significant difference between Promotion 1 and Promotion 3.
- There is a significant difference between Promotion 2 and Promotion 3.

These results indicate that Promotion 2 performs significantly worse than the other promotions, while Promotion 1 and Promotion 3 perform similarly.

Since Promotion 1 has the highest average sales, it is considered the best marketing strategy.

## Final Conclusion

Based on both descriptive analysis and statistical testing, Promotion 1 achieved the highest average sales.

The ANOVA test confirmed that the differences between promotions are statistically significant.

Therefore, Promotion 1 is the most effective marketing strategy and is recommended for implementation.
