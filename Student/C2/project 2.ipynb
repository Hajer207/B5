# Project 2: Student Exam Performance Analysis

## Overview
This project analyzes student performance using academic and lifestyle factors.

## Objective
The goal is to explore how study hours, attendance, and sleep affect exam scores.

## Analysis Approach
We use Exploratory Data Analysis (EDA) to explore patterns and relationships in the data.

## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')
```

## Data Loading

In this step, we load the dataset and preview its structure.

We will check the size of the dataset and display the first few rows.

```python
df = pd.read_csv("StudentPerformanceFactors.csv")

print(df.shape)
df.head()
```

## Quick Data Validation

We perform basic checks to understand the dataset.

This includes checking data types, summary statistics, and missing values.

```python
df.info()
df.describe()
df.isnull().sum()
```

### Insights

The dataset contains 6607 rows and 20 columns.  
All columns have no missing values, which means the data is clean and ready for analysis.

## Distribution of Exam Scores

We explore how exam scores are distributed among students.

This helps us understand the overall performance.

```python
plt.figure(figsize=(6,4))

sns.histplot(df['Exam_Score'], bins=20, color='pink', edgecolor='black')

plt.title("Distribution of Exam Scores", fontsize=14, fontweight='bold')
plt.xlabel("Exam Score")
plt.ylabel("Count")

plt.show()
```

### Insight
Most students scored between 60 and 75.  
This shows that most students have average performance.

## Study Hours vs Exam Score

We analyze the relationship between study hours and exam scores.

This helps us see if studying more leads to better performance.

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df, color='purple')

plt.title("Hours Studied vs Exam Score", fontsize=14, fontweight='bold')

plt.show()
```

### Insight
As study hours increase, exam scores increase.  
This shows a positive relationship between studying and performance.

## Sleep Hours Analysis

We examine how sleep hours are distributed among students.

This helps us understand lifestyle patterns.

```python
plt.figure(figsize=(6,4))

sns.boxplot(x=df['Sleep_Hours'], color='purple')

plt.title("Sleep Hours Distribution")

plt.show()
```

### Insight
Most students sleep between 6 and 8 hours.  
Sleep hours are quite consistent with no major variation.

## Attendance vs Exam Score

We explore how attendance affects exam performance.

This helps us determine if attending classes improves results.

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x='Attendance', y='Exam_Score', data=df, color='purple')

plt.title("Attendance vs Exam Score", fontsize=14, fontweight='bold')

plt.show()
```

### Insight
As attendance increases, exam scores tend to increase.  
This shows that attending classes positively affects performance.

## Conclusion

Study hours and attendance have a strong impact on exam performance.  
Students who study more and attend classes regularly tend to achieve higher scores.  
Other factors like sleep also play a role, but their impact is less significant.
