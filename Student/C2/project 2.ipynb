# Project 2: Student Exam Performance Analysis

## Overview
This project analyzes student performance based on academic and lifestyle factors.

The goal is to explore how different variables such as study hours, attendance, and sleep affect exam scores.

We will perform Exploratory Data Analysis (EDA) to identify patterns, relationships, and key insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')
```

## Data Loading

In this step, we load the dataset and preview its structure.

We will check the size of the dataset and display the first few rows.

```python
df = pd.read_csv("StudentPerformanceFactors.csv")

print(df.shape)
df.head()
```

## Quick Data Validation

We perform basic checks to understand the dataset.

This includes checking data types, summary statistics, and missing values.

```python
df.info()
df.describe()
df.isnull().sum()
```

## Distribution of Exam Scores

We explore how exam scores are distributed among students.

This helps us understand the overall performance.

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Exam_Score'], bins=20)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Count")
plt.show()
```

## Study Hours vs Exam Score

We analyze the relationship between study hours and exam scores.

This helps us see if studying more leads to better performance.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

## Sleep Hours Analysis

We examine how sleep hours are distributed among students.

This helps us understand lifestyle patterns.

```python
sns.boxplot(x=df['Sleep_Hours'])
plt.title("Sleep Hours Distribution")
plt.show()
```

## Attendance vs Exam Score

We explore how attendance affects exam performance.

This helps us determine if attending classes improves results.

```python
sns.scatterplot(x='Attendance', y='Exam_Score', data=df)
plt.title("Attendance vs Exam Score")
plt.show()
```

## Hypothesis

Students who study more hours and attend classes regularly will achieve higher exam scores.

## Key Insights

1. Students with higher study hours tend to achieve higher exam scores.
2. Attendance shows a positive relationship with performance.
3. Most students have average exam scores.
4. Sleep patterns vary among students.
5. Some students perform well even with fewer study hours.
6. There are some outliers in the data.
7. Lifestyle factors influence performance.

## Summary

This analysis shows that student performance is influenced by study habits and attendance.

Students who manage their time effectively tend to perform better.

Future work could include building a predictive model for exam scores.
